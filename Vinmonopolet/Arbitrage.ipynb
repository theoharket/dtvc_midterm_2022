```python
import pandas as pd

df_arbitrage = pd.read_csv("arbitrage_wines.csv")
df_arbitrage["Price Difference"] = (
    df_arbitrage.loc[:, "vivino_price"] - df_arbitrage.loc[:, "vpolet_price"]
)
df_arbitrage["Profit %"] = round((
    df_arbitrage.loc[:, "Price Difference"] - 3
) / df_arbitrage.loc[:, "vpolet_price"] * 100 , 2)

wine_selection = df_arbitrage.sort_values(by="Profit %",ascending=False).drop(columns=['Unnamed: 0',])
wine_selection.reset_index().drop(columns=['index',]).head(10)
```
