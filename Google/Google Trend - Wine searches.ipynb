## Formatting algorithm

This function is designed to be ran after obtaining the google trends from all of the regions above. The function under is assuming you have ran through all of the regions, and as some of the regions doesn't contain any searches on our input (leaving no column on this), there needs to be some formatting of these columns. <br> <br>
Don't run this if you just have read the .csv file in the second cell of the notebook

```python
#Function that formats all necessary parameters in the data set that is sent as a variable to the function
def create_winetable(df, name):
    
    if len(df.columns)!=5:
        red = False
        white = False
        rose = False
        sparkling = False
        pos = 0
        
        #Checking which column(s) is(are) missing
        if df.columns[pos].startswith("red") == False:
            red = True
            df.loc[:, "red wine"] = 0
        pos+=1
        if df.columns[pos].startswith("white") == False:
            white = True
            df.loc[:, "white wine"] = 0
        pos+=1
        if df.columns[pos].startswith("rose") == False:
            rose = True
            df.loc[:, "rose wine"] = 0
        pos+=1
        if df.columns[pos].startswith("sparkling") == False:
            sparkling = True
            df.loc[:, "sparkling wine"] = 0
            
    df = df.reset_index()
    df.loc[:, "region"] = name
    df.columns = ["date", "red wine", "white wine", "sparkling wine", "rose wine", "isPartial", "region"]
    df.loc[:, "total_wines"] = df.loc[:, "red wine"] + df.loc[:,"white wine"] + df.loc[:,"rose wine"] + df.loc[:,"sparkling wine"]
    df.to_csv("%s.csv" %name)
    return df
```

```python
#Creating two lists of every trending data set to make sure that everyone have the same structure and variables to later concatenate them
dataset_list = [Jura, Bordeaux, Castilla_y_Leon, Alsace, Catalonia, Sicily, Tuscany, Loire, Piemonte, Bordeaux, Veneto, Trentino, Rhone, Roussillon, Rioja, Champagne, Burgund, Provence]
name_list = ['Jura', 'Bordeaux', 'Castilla_y_Leon', 'Alsace', 'Catalonia', 'Sicily', 'Tuscany', 'Loire', 'Piemonte', 'Bordeaux', 'Veneto', 'Trentino', 'Rhone', 'Roussillon', 'Rioja', 'Champagne', 'Burgund', 'Provence']

#Temporary list to store the dataframes returned of the above function
dataframe_list = []

#Iterates through every dataframe of the trends, formatting the tables, returning a formatted data frame
for i in range(len(dataset_list)):
    df = create_winetable(dataset_list[i], name_list[i])
    dataframe_list.append(df)

#Creating a combined dataset of the above dataframe list
df_trend_region = pd.concat(dataframe_list)

#Storing it in a .csv file
df_trend_region.to_csv("combined_regions_google_trend.csv")
```

## Visualization of trending total searches - Top 10 regions

```python
#Reading the data from a .csv file rather than running through every cell of this notebook
df_trend_region = pd.read_csv("combined_regions_google_trend.csv")

#Getting top 10
top_regions = df_trend_region.groupby("region")["total_wines"].sum().sort_values(ascending = False)
top_top5_regions = top_regions.index[0:5]
top_next5_regions = top_regions.index[5:10]

#Creating data frame of the top 5 regions for plotting with a thicker line
df_region_top5 = df_trend_region[df_trend_region["region"].isin(top_top5_regions)]

#Creating data frame with the next 5 regions for ploitting with a thinner line than top 5
df_region_next5 = df_trend_region[df_trend_region["region"].isin(top_next5_regions)]
```

```python
#Setting figure size
plt.figure(figsize = (20,8))

#Plotting the figures
sns.lineplot(x = "date", y = "total_wines", hue = "region", data = df_region_top5, linewidth = 5)
sns.lineplot(x = "date", y = "total_wines", hue = "region", data = df_region_next5, linewidth = 0.5)

#Setting title
plt.title("Total searches for wine per region", fontsize = 25, fontfamily = "Georgia")

#Storing legend in a variable
leg = plt.legend()
tmp = 0

#Setting linewidth for the first 5
for line in range(5): leg.get_lines()[line].set_linewidth(5)

#Changing font of legend
plt.setp(leg.texts, family='Georgia', fontsize = 15)

#Setting xticks for the graph
pos = [i for i in range(0, 52, 4)]
mnth = ["10-2021", "11-2021", "12-2021", "01-2022", "02-2022", "03-2022", "04-2022", "05-2022", "06-2022", "07-2022", "08-2022", "09-2022", "10-2022"]
plt.gca().set(xticks = pos, xticklabels=mnth)

#Setting xlabel and ylabel
plt.xlabel("Date", fontsize = 15, fontfamily = "Georgia")
plt.ylabel("Total searches - Top 10", fontsize = 15, fontfamily = "Georgia")

#Saving the figure
plt.savefig("trend.jpeg",dpi=600)

#Displaying the figure
plt.show()
```

## Visualisation on trending searches in top 5 regions


```python
#Creating data frame of all the regions
df_bordeaux = df_trend_region[df_trend_region["region"] == "Bordeaux"]
df_sicily = df_trend_region[df_trend_region["region"] == "Sicily"]
df_loire = df_trend_region[df_trend_region["region"] == "Loire"]
df_champagne = df_trend_region[df_trend_region["region"] == "Champagne"]
df_provence = df_trend_region[df_trend_region["region"] == "Provence"]
```

```python
#Storing the data frames in a list
df_list = []

#Function that takes a data frame as a variable, and returns a formatted data frame with relative percents on the wines
def create_dataset(df):

    gb = df.groupby("region")[["red wine", "white wine", "sparkling wine", "rose wine", "total_wines"]].sum()
    gb = gb.reset_index()

    gb.loc[:, "red wine"] = float(gb.loc[:, "red wine"]/gb.loc[:, "total_wines"])*100
    gb.loc[:, "white wine"] = float(gb.loc[:, "white wine"]/gb.loc[:, "total_wines"])*100
    gb.loc[:, "sparkling wine"] = float(gb.loc[:, "sparkling wine"]/gb.loc[:, "total_wines"])*100
    gb.loc[:, "rose wine"] = float(gb.loc[:, "rose wine"]/gb.loc[:, "total_wines"])*100

    return gb

#Running through all of the original data frames that needs cleaning
for data in [df_bordeaux, df_sicily, df_loire, df_champagne, df_provence]: df_list.append(create_dataset(data))
```
