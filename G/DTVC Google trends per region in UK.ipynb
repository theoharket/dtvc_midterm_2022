```python
def create_winetable(df, name):
    df= df.reset_index()
    df.loc[:, "region"] = name
    df.columns = ["date", "red wine", "white wine", "sparkling wine", "rose wine", "isPartial", "region"]
    df.loc[:, "total_wines"] = df.loc[:, "red wine"] + df.loc[:,"white wine"] + df.loc[:,"rose wine"] + df.loc[:,"sparkling wine"]
    return df
```
